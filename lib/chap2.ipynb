# Chapter 1. Module

Moving code to a module is a good way to hide low level implementation details.

Notebook executes module's code, explains methodology, presents and describes results etc.

## Paths

Avoid hard-coding absolute paths, this is not portable. Use relative paths.

Since code sits in different places, it is a good idea to stick to project root as working directory. Then all code is imported from `lib.name_of_module`, all data are accessed from `data/...` etc.

````python
%cd ..
````

## Simplest approach: Everything happens in the script

It can be as simple as
```python
from lib import chap1
```
> Output of chap1.py is displayed here.

Import statement executes code of `chap1.py`. All processing is hidden, and notebook only displays results.

## More advanced: Interface vs implementation

Decide what's important to show, and what can be hidden. This applies to both results and processing.

````python
from lib import chap2
%matplotlib inline

chap2.df.plot()
````

````python
chap2.summarize()
````

````python
chap2.print_answer()
````

## Underscore variables

Underscore variables defined in the module are pseudo-private. They don't show up on tab completion. But they are still accessible. This can be useful for interactive development.

````python
chap2._data[:10]
````
